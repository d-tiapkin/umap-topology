```python

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 
import gudhi
%matplotlib notebook
```

```python
def load_points(file):
    return np.genfromtxt(file)
```

```python
def draw_R2(points):
    fig,ax = plt.subplots(1, figsize=(10,10))
    ax.scatter(points[:,0], points[:,1], c='blue')
    plt.show()
```

```python
def draw_R3(points):
    fig = plt.figure(figsize=(10,10))
    ax = fig.gca(projection='3d')
    ax.scatter(points[:,0], points[:,1], points[:,2], c='red')
    plt.show()
```

```python
def draw_persistence(file, figsize):
    fig,ax = plt.subplots(1, figsize=figsize)
    gudhi.plot_persistence_barcode(persistence_file=file, legend=True).show()
```

```python
fig,ax = plt.subplots(1, figsize=(18,180))
gudhi.plot_persistence_barcode(persistence_file="/home/unkoll/UMAP-git/umap-topology/filtration/SO3_Z3.dgm", legend=True)
plt.show()
```
